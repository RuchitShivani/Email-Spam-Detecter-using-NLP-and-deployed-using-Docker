```python
import pandas as pd
import re
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

# Step 1: Load dataset
df = pd.read_csv('spam.csv', encoding='ISO-8859-1', usecols=['v1', 'v2'])
df.columns = ['label', 'message']

# Step 2: Encode labels (ham=0, spam=1)
df['label'] = df['label'].map({'ham': 0, 'spam': 1})

# Step 3: Basic text preprocessing (removed unwanted characters but preserved relevant words)
def clean_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'\d+', '', text)  # Remove digits
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    text = re.sub(r'\s+', ' ', text)  # Remove extra spaces
    return text.strip()

df['cleaned'] = df['message'].apply(clean_text)

# Step 4: Feature extraction
vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)
X = vectorizer.fit_transform(df['cleaned'])
y = df['label']

# Step 5: Train-test split and model training
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

model = MultinomialNB()
model.fit(X_train, y_train)

# Step 6: Evaluation
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))

# Step 7: Prediction function
def predict_spam(text):
    cleaned = clean_text(text)
    vec = vectorizer.transform([cleaned])
    return "Spam" if model.predict(vec)[0] == 1 else "Ham (Not Spam)"

# Example usage
print(predict_spam("Congratulations! You've won a free cruise trip. Click here."))
print(predict_spam("Let's catch up at 5 PM. See you!"))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report, precision_recall_curve, average_precision_score
import numpy as np
import pandas as pd

# 1. Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Ham', 'Spam'], yticklabels=['Ham', 'Spam'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# 2. Classification Report as Table
report = classification_report(y_test, y_pred, target_names=['Ham', 'Spam'], output_dict=True)
report_df = pd.DataFrame(report).transpose()
plt.figure(figsize=(6,3))
sns.heatmap(report_df.iloc[:-1, :-1], annot=True, cmap='YlGnBu')
plt.title('Classification Report')
plt.show()

# 3. Precision-Recall Curve
y_score = model.predict_proba(X_test)[:,1]
precision, recall, _ = precision_recall_curve(y_test, y_score)
ap_score = average_precision_score(y_test, y_score)

plt.figure(figsize=(6,4))
plt.plot(recall, precision, marker='.', label=f'AP={ap_score:.2f}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend()
plt.show()

# 4. Top 20 Spam Indicator Features
feature_names = vectorizer.get_feature_names_out()
coefs = model.feature_log_prob_[1] - model.feature_log_prob_[0]
top_spam_idx = np.argsort(coefs)[-20:]
top_spam_terms = feature_names[top_spam_idx]
top_spam_scores = coefs[top_spam_idx]

plt.figure(figsize=(8,6))
sns.barplot(x=top_spam_scores, y=top_spam_terms, palette='Reds_r')
plt.title('Top 20 Spam Indicator Words')
plt.xlabel('Importance (Log Probability Difference)')
plt.ylabel('Word')
plt.show()
```

```python
import pandas as pd
import nltk
import string
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

# Download NLTK resources (only needs to be run once)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Sample dataset
data = {
    'text': [
        # Positive
        "I love this product! It's amazing.",
        "Excellent quality and fast delivery.",
        "Very satisfied with the service provided.",
        "Absolutely fantastic! Highly recommend it.",
        "Five stars! Will definitely buy again.",
        "The color and fit were perfect!",
        "Amazing experience overall. Kudos to the team!",
        "It works flawlessly and exceeded my expectations.",
        "Best purchase of the year so far.",
        "Impressed with the performance and packaging.",
        # Negative
        "Terrible experience. Will never buy again!",
        "Worst purchase I've made. Very disappointed.",
        "The item broke after two days of use.",
        "Customer service was rude and unhelpful.",
        "It didn't match the description at all.",
        "Poor quality and late delivery.",
        "Very unsatisfied. Will not recommend to anyone.",
        "A complete waste of money and time.",
        "Defective item received. Very upset.",
        "Shipping was delayed and item was damaged.",
        # Neutral
        "The item was okay, not great but not bad.",
        "Its an average product, does the job.",
        "Neutral feelings about this purchase.",
        "Neither satisfied nor disappointed.",
        "Just what I expected. Nothing more, nothing less.",
        "The product is decent for the price.",
        "Not amazing, not terrible  just okay.",
        "The service was fine, nothing exceptional.",
        "Regular product. Works as expected.",
        "Functionality is acceptable, but not impressive."
    ],
    'sentiment': [
        'positive', 'positive', 'positive', 'positive', 'positive',
        'positive', 'positive', 'positive', 'positive', 'positive',
        'negative', 'negative', 'negative', 'negative', 'negative',
        'negative', 'negative', 'negative', 'negative', 'negative',
        'neutral', 'neutral', 'neutral', 'neutral', 'neutral',
        'neutral', 'neutral', 'neutral', 'neutral', 'neutral'
    ],
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Preprocessing function
def preprocess_text(text):
    tokens = text.lower().split()
    lemmatizer = WordNetLemmatizer()
    stop_words = set(stopwords.words('english'))
    cleaned = [
        lemmatizer.lemmatize(word) for word in tokens
        if word not in stop_words and word not in string.punctuation
    ]
    return ' '.join(cleaned)

# Apply preprocessing
df['cleaned_text'] = df['text'].apply(preprocess_text)

# Feature extraction
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df['cleaned_text'])
y = df['sentiment']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8, random_state=42)

# Model training
model = LogisticRegression()
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
import os
import string
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# optional: simple stopwords list (you can expand this)
STOPWORDS = set("""
a an the and or but if while of at by for with about against between into through during before after to from in on
""".split())

# Step 1: Load Documents
def load_documents(folder_path):
    documents = []
    filenames = []
    for filename in os.listdir(folder_path):
        if filename.endswith(".txt"):
            with open(os.path.join(folder_path, filename), 'r', encoding='utf-8') as file:
                documents.append(file.read())
                filenames.append(filename)
    return documents, filenames

# Step 2: Preprocess Text (basic, without nltk)
def preprocess(text):
    text = text.lower()
    text = text.translate(str.maketrans("", "", string.punctuation))
    words = text.split()
    filtered = [word for word in words if word not in STOPWORDS]
    return " ".join(filtered)

# Step 3: Vectorize Documents
def vectorize(documents):
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(documents)
    return vectorizer, tfidf_matrix

# Step 4: Search Function
def search(query, vectorizer, tfidf_matrix, raw_docs, filenames, top_n=3):
    query_vec = vectorizer.transform([preprocess(query)])
    similarities = cosine_similarity(query_vec, tfidf_matrix).flatten()
    ranked_indices = similarities.argsort()[:-top_n-1:-1]

    print("\nTop Results:\n")
    for index in ranked_indices:
        print(f"Score: {similarities[index]:.4f} | File: {filenames[index]}")
        print(raw_docs[index][:300].strip(), "\n---\n")


# Step 5: Main Execution
def main():
    folder_path = "/content/Ruchit,r"  # Folder with .txt files
    print("Loading documents...")
    raw_docs, filenames = load_documents(folder_path)

    print("Preprocessing...")
    processed_docs = [preprocess(doc) for doc in raw_docs]

    print("Vectorizing...")
    vectorizer, tfidf_matrix = vectorize(processed_docs)

    while True:
        query = input("\nEnter your search query (or type 'exit' to quit):\n")
        if query.lower() == 'exit':
            break
        search(query, vectorizer, tfidf_matrix, raw_docs, filenames)

if __name__ == "__main__":
    main()
```
